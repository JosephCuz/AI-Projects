```
# This is a helper function to fix the title and will be paired with reorder_title(text)
def remove_parentheses_text(text):
  items = text.split("(")
  return items[0][:-1]
```

```
# this gets the year from the 'title' column as getting it from the 'release date' column in the
# 'items' dataset led to some inconsistancies
def get_year_from_title(text):
  items = re.findall("\(\d{4}\)", text)
  # a title did not have an associated year
  if len(items) == 0:
    return 0
  year_str = items[0]
  year_itm = re.findall("\d+", year_str)
  return year_itm[0]
```

```
# titles were named Avengers, The. This function replaces that with The Avengers
def reorder_title(text):
  items = text.split(", ")
  items.reverse()
  return " ".join(items)
```

```
# Just makes things easier
def apply_functions(fn_1, fn_2, text):
  return fn_2(fn_1(text))
```

```
#now simple model
global_mean = y_trainU1.mean()
from implicit.als import AlternatingLeastSquares
from scipy.sparse import coo_matrix

# 1.1  build user-item matrix (rows = users, cols = movies, data = ratings)
rows = trainU1["userId"].values
cols = trainU1["movieId"].values
data = trainU1["rating"].values.astype(np.float32)

R = coo_matrix((data, (rows, cols)))

als = AlternatingLeastSquares(
    factors       = 50,     # latent dims
    regularization= 0.1,
    iterations    = 20,
    use_gpu       = False,  # True if you have CUDA
    random_state  = 42
)
als.fit(R)                 # ≈ 3–4 s on ML-100K

def als_predict(u, m):
    if u < als.user_factors.shape[0] and m < als.item_factors.shape[0]:
        return als.user_factors[u] @ als.item_factors[m]
    return global_mean                       # cold-start fallback

val_preds = [als_predict(u, m)
             for u, m in zip(X_valU1["userId"], X_valU1["movieId"])]

from sklearn.linear_model import LinearRegression

# --- 1.1  get ALS scores on the *training* rows -------------
train_scores = [als_predict(u, m)
                for u, m in zip(X_trainU1["userId"], X_trainU1["movieId"])]
train_y      = y_trainU1.values

# --- 1.2  learn a linear map:  score → rating ---------------
scaler = LinearRegression().fit(np.array(train_scores).reshape(-1,1), train_y)

# --- 1.3  apply map to the validation set ------------------
val_scores = [als_predict(u, m)
              for u, m in zip(X_valU1["userId"], X_valU1["movieId"])]
val_preds  = scaler.predict(np.array(val_scores).reshape(-1,1))

rmse_val = mean_squared_error(y_valU1, val_preds)
rmse_val = np.sqrt(rmse_val)
print(f"Calibrated ALS RMSE (val): {rmse_val:.4f}")

```

```
#SGD implementation

import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error

# ------------------------------------------------------------------
# 1  Prepare the data  (DataFrame with userId, movieId, rating columns)
# ------------------------------------------------------------------

ratings = trainU1[["userId","movieId","rating"]]      # keep only what we need
n_users = users["userId"].max() + 1
n_items = items["movieId"].max() + 1

train = ratings.sample(frac=0.8, random_state=42)
val   = ratings.drop(train.index)

# ------------------------------------------------------------------
# 2  Hyper-parameters
# ------------------------------------------------------------------
k      = 50        # latent dimension
lr     = 0.01      # learning-rate γ
reg    = 0.05      # λ in the loss
epochs = 20

# ------------------------------------------------------------------
# 3  Initialise latent matrices and bias terms (optional)
# ------------------------------------------------------------------
P  = 0.1 * np.random.randn(n_users, k)   # user factors
Q  = 0.1 * np.random.randn(n_items, k)   # item factors

# ------------------------------------------------------------------
# 4  SGD loop  (the update equations from the slide)
# ------------------------------------------------------------------
train_mat = train[["userId","movieId","rating"]].values
for epoch in range(epochs):
    np.random.shuffle(train_mat)          # good practice for SGD
    for u, i, r in train_mat:
        pred = P[u] @ Q[i]                # p_i^T q_j
        err  = r - pred                   # (r_ij − p_i^T q_j)

        # gradient updates (note the 2·γ from the formula)
        P[u] += lr * ( 2*err * Q[i] - reg * P[u] )
        Q[i] += lr * ( 2*err * P[u] - reg * Q[i] )

    # quick progress check every few epochs
    if (epoch+1) % 5 == 0:
        preds = [P[u] @ Q[i] for u,i,r in train_mat]
        rmse  = mean_squared_error(train_mat[:,2], preds)
        rmse = np.sqrt(rmse)
        print(f"Epoch {epoch+1}/{epochs}  –  train RMSE: {rmse:.3f}")

# ------------------------------------------------------------------
# 5  Evaluate on the validation split
# ------------------------------------------------------------------
val_preds = [P[u] @ Q[i] for u,i,r in val.values]
val_rmse  = mean_squared_error(val["rating"], val_preds)
val_rmse = np.sqrt(val_rmse)
print(f"\nValidation RMSE: {val_rmse:.3f}")

```

```
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import pandas as pd, numpy as np

# ─── 0.  helper vocabularies ──────────────────────────────────────────────

#map each zip to an integer which will be used for emebeddings
zip_list  = sorted(users.zip.unique())
zip2idx   = {z: i for i, z in enumerate(zip_list)}
UNK_ZIP   = len(zip2idx)  #unseen codes
n_zip_emb = UNK_ZIP + 1 #emebedding input dim (num of potential zipcodes)


#map each occupation to a integer to be used for embeddings
occ_list   = sorted(users.occupation.unique())
occ2idx    = {occ: i for i, occ in enumerate(occ_list)}
UNK_OCC    = len(occ2idx)                       #for unknown/unseen
n_occ_emb  = UNK_OCC + 1                        #Embedding input dim

#map gender to 0/1
gender2idx = {"F": 0, "M": 1}

#numeric genre matrix (for shape n_items × 19), will map movie to list of genre flags [0, 1, 0, 1,...]
genre_cols        = [c for c in item_feats.columns if c.startswith("genre_")]
item_genre_matrix = item_feats[genre_cols].astype("float32")

# ─── 1. input sizes ───────────────────────────────────────────────────────
n_users  = users.userId.max()  + 1  #userId input should have a slot for every potential user, this allows distinction among user cases
n_items  = items.movieId.max() + 1 #same for movies
n_genders = 2
n_genres  = len(genre_cols)

# ─── 2. model definition ──────────────────────────────────────────────────

# ─── ZIP code input embedding ───────────────────────────────
#define input tensor
inp_zip  = keras.Input(shape=(), dtype="int32", name="zip_idx") #will be a single integer value
emb_zip  = layers.Embedding(n_zip_emb, 8, name="zip_emb")(inp_zip) #emebedding layer to map a zip code to 8-d vector, connect emb layer to input layer, this feeds inp layer result (idx) into emb layer (idx -> 8d emb)
emb_zip  = layers.Flatten()(emb_zip)          #connect emb layer to flatten layer, results in size (batch_size, 8)


# ID embeddings
inp_user = keras.Input(shape=(), dtype="int32", name="userId") #input is int value
inp_item = keras.Input(shape=(), dtype="int32", name="movieId")
emb_user = layers.Embedding(n_users, 32)(inp_user) #connect inp layer to embedding layer (id -> 32D emb)
emb_item = layers.Embedding(n_items, 32)(inp_item)

# user metadata
inp_gender = keras.Input(shape=(n_genders,), dtype="float32", name="gender_1hot") #input will be vector of flags 2 vals for gender
inp_occ    = keras.Input(shape=(), dtype="int32",   name="occupation_idx")   #input will be single value representing given occupation
inp_age    = keras.Input(shape=(1,), dtype="float32", name="age_norm") #input will be age value (1d array, cuz numerical and concatenating later)

emb_occ  = layers.Embedding(input_dim=n_occ_emb, output_dim=8, name="occ_emb")(inp_occ) #connect input to emb, (occVal -> 8d emb)
emb_occ  = layers.Flatten()(emb_occ)

dense_gender = layers.Dense(8, activation="relu")(inp_gender) #embedds 2d gender input to 8d representation

# movie metadata (genres)
inp_genres = keras.Input(shape=(n_genres,), dtype="float32", name="genre_1hot") #input is genre flags
dense_gen  = layers.Dense(16, activation="relu")(inp_genres) #embedds genre flags to 16d representation

# ── concatenate & tower ───────────────────────────────────────────────────
x = layers.Concatenate()([ #concatenate all our input layers together, really we are concatenating their outputs together
    layers.Flatten()(emb_user), #needs flattening
    layers.Flatten()(emb_item),
    dense_gender,
    emb_occ,
    emb_zip,
    inp_age,
    dense_gen
])

#define our deep layers after inputs and emebeddings
x = layers.Dense(128, activation="relu")(x)
x = layers.Dropout(0.3)(x)
x = layers.Dense(64,  activation="relu")(x)
x = layers.Dropout(0.3)(x)
x = layers.Dense(64,  activation="relu")(x)
x = layers.Dropout(0.3)(x)
x = layers.Dense(32,  activation="relu")(x)
out = layers.Dense(1,  activation="linear")(x)

#create the model
model = keras.Model(
    inputs=[inp_user, inp_item, inp_gender, inp_occ, inp_zip, inp_age, inp_genres], #our list of input layers, defined above (these lead into the embeddings that come out in the concatenated layer)
    outputs=out
)
model.compile(optimizer=keras.optimizers.Adam(1e-3),
              loss="mse",
              metrics=[keras.metrics.RootMeanSquaredError()])
model.summary()

# ─── 3. batch builder ─────────────────────────────────────────────────────
#need helper function to organize input data to expected format for the model
def make_batch(df: pd.DataFrame):
    uids = df.userId.values.astype("int32") #extract userids
    mids = df.movieId.values.astype("int32") #extract movieids

    # gender 1-hot (F → [1,0], M → [0,1])
    gids = df.gender.map(gender2idx).astype("int32").values #map genders to one hot
    gender_ohs = tf.one_hot(gids, depth=2).numpy().astype("float32")

    # occupation scalar IDs, unseen → UNK_OCC
    oids = df.occupation.map(occ2idx).fillna(UNK_OCC).astype("int32").values #map occupations to ints

    zids = df.zip.map(zip2idx).fillna(UNK_ZIP).astype("int32").values #map zip code to int

    #normalize ages
    age_norm = ((df.age - users.age.mean()) / users.age.std())\
               .astype("float32").values.reshape(-1, 1)

    genres = item_genre_matrix.loc[df.movieId].values.astype("float32") #lookup corresponding genre flags for each movie

    return { #return input transformations
        "userId":         uids,
        "movieId":        mids,
        "gender_1hot":    gender_ohs,
        "occupation_idx": oids,
        "zip_idx":        zids,
        "age_norm":       age_norm,
        "genre_1hot":     genres,
    }

# ─── 4. data preparation ──────────────────────────────────────────────────
#merge rating to corresponding user and movie data
train_df = trainU1.merge(users, on="userId").merge(items, on="movieId")
val_df   = valU1.merge(users,  on="userId").merge(items,  on="movieId")
test_df  = testU1.merge(users, on="userId").merge(items, on="movieId")

#feed into helper to format for model input and extract labels
train_data, train_labels = make_batch(train_df), train_df.rating.astype("float32")
val_data,   val_labels   = make_batch(val_df),   val_df.rating.astype("float32")
test_data,  test_labels  = make_batch(test_df),  test_df.rating.astype("float32")

# ─── 5. training ──────────────────────────────────────────────────────────
model.fit(
    x=train_data,
    y=train_labels,
    validation_data=(val_data, val_labels),
    epochs=20,
    batch_size=512,
    verbose=2
)

test_loss, test_rmse = model.evaluate(test_data, test_labels, verbose=0)
print(f"Test RMSE: {test_rmse:.4f}")
```

Essentially a 5-layer deep NN hybrid recommender that takes everything into account to predict the 1-5 star rating that a given user will have for a movie, based on user data (occupation, users rating history, gender, etc.) and movie data (genre). The model embeds all categorical information, the resulting embeddings are all concatenated together to one long layer, which is then fed into a series of dense and dropout layers to calculate the rating. During training, the weights in both the dense layers and the embedding layers are optimized


Implementation notes:
Adding in an embedding for occupations after mapping them to indx resulted in a noticable decrease in RMSE (~.02). Testing out different fully connected layer configurations was essential to find the ideal combination of nuerons, layers, and dropout layers.
The best result we are seeing on the test split is ~0.95, and it consistently gets between .95 and .97 which is a solid improvement over the previous non-hybrid models.
Adding zip code information did not make a significant difference, disappointing.
